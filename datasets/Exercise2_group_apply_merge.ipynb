```python
covid2 = covid.dropna()
covid2
```

```python
count_covid = covid2.groupby('province_of_isolation').count()
count_covid.sort_values(by=["No."],ascending=False)
```

```python
# create a function
def gender_to_numeric(x):
    if x == 'ชาย' or x == 'นาย':
        return 1
    if x == 'หญิง':
        return 0
```

```python
covid2['sex_n'] = covid2['sex'].apply(gender_to_numeric)
covid2
```

```python
a = covid2.groupby('sex').sex_n.sum() / covid2.sex_n.count() * 100 
a.sort_values(ascending = False)
```

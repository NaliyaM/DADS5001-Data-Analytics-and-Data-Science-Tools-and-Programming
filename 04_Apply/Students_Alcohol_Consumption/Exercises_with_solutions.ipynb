### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
```

```python
majority2 = lambda x: True if x > 17 else False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority2)
stud_alcoh.head()
```
